```python
#Function to remove punctuation
import string

def remove_punctuation(text):
    return text.translate(str.maketrans('', '', string.punctuation))

df['problem'] = df['problem'].apply(remove_punctuation)
df['solution'] = df['solution'].apply(remove_punctuation)
```

```python
#Removing stop words as they do not add any value to the analysis

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Set of English stop words
stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    # Tokenize the text to words
    words = word_tokenize(text)
    # Remove stop words and return the cleaned text
    return ' '.join([word for word in words if word not in stop_words])

# Apply the function to your DataFrame columns
df['problem_no_stopwords'] = df['problem'].apply(remove_stopwords)
df['solution_no_stopwords'] = df['solution'].apply(remove_stopwords)
```

```python
import re

# Function to check for URLs
def contains_url(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return bool(url_pattern.search(text))

# Function to check for HTML tags
def contains_html_tags(text):
    html_pattern = re.compile(r'<[^>]+>')
    return bool(html_pattern.search(text))

# Apply these functions to your DataFrame
df['contains_url'] = df['problem_no_stopwords'].apply(contains_url)
df['contains_html'] = df['solution_no_stopwords'].apply(contains_html_tags)
```

```python
from textblob import TextBlob

# Function for keyword extraction (here, just extracting nouns as an example)
def extract_keywords(text):
    blob = TextBlob(text)
    return [word for word, tag in blob.tags if tag in ['NN', 'NNS']]

# Function for sentiment analysis
def get_sentiment(text):
    blob = TextBlob(text)
    return blob.sentiment.polarity

# Apply keyword extraction and sentiment analysis
df['problem_keywords'] = df['problem_clean'].apply(extract_keywords)
df['solution_keywords'] = df['solution_clean'].apply(extract_keywords)
df['problem_sentiment'] = df['problem_clean'].apply(get_sentiment)
df['solution_sentiment'] = df['solution_clean'].apply(get_sentiment)
```

```python
def tokenize_text(text):
    # Check if the text is a string
    if isinstance(text, str):
        return word_tokenize(text)
    else:
        # Return an empty list or some placeholder if text is not a string
        return []

# Apply tokenization to the 'problem' and 'solution' columns
df['problem_tokens'] = df['problem'].apply(tokenize_text)
df['solution_tokens'] = df['solution'].apply(tokenize_text)

# Display the DataFrame to verify the results
df[['problem_tokens', 'solution_tokens']].head()
```

```python
def score_solution(row):
    print("\nSolution:", row['solution'])
    scores = {}
    for kpi in ["Innovation and Feasibility", "Resource Efficiency", "Environmental Impact", "Scalability and Adaptability"]:
        while True:
            try:
                score = int(input(f"Enter score for {kpi} (1-5): "))
                if 1 <= score <= 5:
                    scores[kpi] = score
                    break
                else:
                    print("Score must be between 1 and 5.")
            except ValueError:
                print("Invalid input. Please enter a numeric value.")
    return pd.Series(scores)
```

```python
from transformers import AutoModelForSequenceClassification, Trainer, TrainingArguments


# Load your dataset
dataset = pd.copy(df)

# Load the pre-trained model
model = AutoModelForSequenceClassification.from_pretrained('mistralai/Mistral-7B-v0.1', num_labels=4)

# Define training arguments
training_args = TrainingArguments(
    output_dir='./results',
    num_train_epochs=3,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=64,
    warmup_steps=500,
    weight_decay=0.01,
    logging_dir='./logs',
)

# Initialize Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=dataset['train'],
    eval_dataset=dataset['validation']
)

# Train the model
trainer.train()
```
